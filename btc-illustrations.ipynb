```python
import matplotlib.pyplot as plt
import data_preprocessing
import nn_helpers
import pandas as pd
import numpy as np
```

```python
main_df = data_preprocessing.get_data('MSFT')
main_df.head()
```

```python
plt.plot(main_df['Date'][-10:-1], main_df['Close'][-10:-1], label='prev')
plt.plot(main_df['Date'][-2:], main_df['Close'][-2:], label='current')

plt.xticks(rotation=90)

plt.title('MSFT')
plt.xlabel('time')
plt.ylabel('price in USD')

plt.legend()
# plt.show()
plt.savefig(f'illustrations/MSFT.svg', format='svg')
# plt.savefig(f'illustrations/BTC-USD-{main_df["Date"].min().year}_{main_df["Date"].min().month}-{main_df["Date"].max().year}_{main_df["Date"].max().month}.png', format='png')
```

```python
date_range = main_df['Date'][-10:-1]
plt.plot(date_range, np.random.rand(len(date_range)), 'g', label='positive')
plt.plot(date_range, np.random.rand(len(date_range)), 'r', label='negative')
plt.plot(date_range, np.random.rand(len(date_range)), 'b', label='neutral')

plt.xticks(rotation=90)

plt.title('News')
plt.xlabel('time')
plt.ylabel('sentiment coef')

plt.legend()
plt.savefig(f'illustrations/news.svg', format='svg')
plt.show()
```

```python
cut_df = main_df[main_df['Date'] > '2019-01-01']

train_perc = 0.8
len_train = int(len(cut_df) * train_perc)

plt.plot(cut_df[:len_train]['Date'], cut_df[:len_train]['Close'], label='train')
plt.plot(cut_df[len_train:]['Date'], cut_df[len_train:]['Close'], label='test')
plt.legend(loc='upper left')
plt.xticks(rotation=90)
plt.savefig('illustrations/btc-cut-train_test_split.png', format='png')
```
